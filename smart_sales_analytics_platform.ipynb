```python
import matplotlib.pyplot as plt

df.groupby("day")["total_bill"].sum().plot(kind="bar")
plt.title("Sales by Day")
plt.show()
```

Endi qaysi kun Restoran ko'proq pul topishini ko'rib chiqamiz:

```python
top_day = df.groupby("day")["total_bill"].sum().idxmax()
```

```python
top_day_value = df.groupby("day")["total_bill"].sum().max()
```

```python
print("Business Insights:")
```

```python
print(f"- {top_day} is the most profitable day with total sales of {top_day_value:.2f}")
```
