```python
data1 = np.load('GM_Cars/gm_2020.npz')
data2 = np.load('GM_Cars/gm_2021.npz')

models1 = data1['models']
qnts1 = data1['qnts']

models2 = data2['models']
qnts2 = data2['qnts']

mask1 = (models1=='Damas') | (models1=='Nexia') | (models1=='Cobalt') | (models1=='Gentra')
mask2 = (models2=='Damas') | (models2=='Nexia') | (models2=='Cobalt') | (models2=='Gentra')

qnts1[mask1], qnts2[mask2]
```

4-Topshiriq

```python
selected1 = qnts1[mask1]
selected2 = qnts2[mask2]

(selected1 / qnts1.sum()) * 100, (selected2 / qnts2.sum()) * 100
```
